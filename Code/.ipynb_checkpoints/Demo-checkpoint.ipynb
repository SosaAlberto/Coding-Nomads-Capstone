### One hot encoding
One hot encoding consit of transforming a feature that contains different elements in it (categorical data, in this case the genre column) into different features containing only 1 element that could be either true (1) or false (0) (binary vector, meaning is 1 or 0).At the end if our original genre feature had n different genres, the resulting dataframe will have n new features representing an element of the original feature.

EX:

$$\begin{bmatrix}
 a\\
 b\\
 c\\
a
\end{bmatrix}
\to 
\begin{bmatrix} 
\hline
a & b & c \\
 1 & 0 & 0 \\
 0 & 1 & 0 \\
 0 & 0 & 1 \\
 1 & 0 & 0 \\
\end{bmatrix}
$$

**A dataframe with 1 feature and 4 entries transforms into a dataframe with 3 features where the new columns represent a, b and c.**

The reason we use one hot encoding is because it is easier to manage a dataframe where its features represent only one thing. In addition, now we have numerical data representing categorical data (ML algorithms need numbers) instead of strings of different categories.

Create a column per genre using df.at[row,new_column]

```python
def one_hot_encoding(df):
    ##Get pd.series per row and iterate through index
    for index,row in df.iterrows():
        for genre in list(row['genres']):
        #if genre in list create column and equal that row to one
            df.at[index, genre] = 1.0
    df=df.fillna(0)
    
    #Drop original genres column and extra column created
    df.drop(['genres'],axis=1,inplace=True)
    df.drop(['(no','listed)'],axis=1,inplace=True)
    return df
```

# Filtering algorithms



To recommend movies to a given user, it is necessary to find a way to filter down all the set of movies to a subset of movies the user may find interesting.
In order to accomplish that, we will implement to different methods:
* Collaborative filtering
* Content filtering

In this notebook, there will be 4 different filtering algorithms used:
* User-user: Calculate similarity between users to make recommendations based on similar users, can recommend really varied items.(Collaborative)
* User-item: Calculate preferences of users to items with certain charactersitics, recommend only similar items.
* Item-item: Calculate similarity between items and recommend items with similar features.(Content)
* Item-item 2 (synopsis): Calculate similarity between synopsis and make recommendations base on that


### Collaborative filtering

**Collaborative Filtering uses similarity between users to determine what to recommend to the given user.**

## Pearson Correlation Coefficient
The algorithm calculates how much the vectors correlate between two users. If the trends are similar the value will be close to 1, if there is no relation at all the value will be closed to 0 and if there is negative relation the value will be close to -1.

In the case of recommendation systems, and in particular this approach which consist on user-user relation, the algorithm will filter out the users that are not similar to the target and keeping a list of similar users (the list before was a list of user that have watched the same movies or a fraction of them)

**Note: correlation measures how close two vectors are to each other (measures the distance,trend etc)** 


### Mathematical representation


$ \frac{\sum \limits _{i=0} ^{n}(rating_i -mean)(rating _j - mean)} {\sqrt{\sum \limits _{i=0} ^{n}(rating_i -mean)^2}\sqrt{\sum\limits _{i=0} ^{n}(rating_j -mean)^2}} $



```python
def pearson_corr(user_group,input_df):
    pearsonCorrelationDict = {}

   #name is user id and group is movies that user has watched
    for name, group in userSubsetGroup:
       
        group = group.sort_values(by='movieId')
        input_df = input_df.sort_values(by='movieId')
        #Get the N for the formula
        nRatings = len(group)
        #Get the review scores for the movies that they both have in common
        temp_df = input_df[input_df['movieId'].isin(group['movieId'].tolist())]
        tempRatingList = temp_df['rating'].tolist()
        
        tempGroupList = group['rating'].tolist()
        user_average=temp_df['rating'].mean()
        crew_average=group['rating'].mean()
        num=sum((i-user_average)*(j-crew_average) for i,j in zip(tempRatingList,tempGroupList))
        magnitude_x=sqrt(sum((i-user_average)**2 for i in tempRatingList))
        magnitude_y=sqrt(sum((i-crew_average)**2 for i in tempGroupList))
        den=magnitude_x*magnitude_y

        if num!=0 and den!=0:
            pearsonCorrelationDict[name] = num/den
        else:

            pearsonCorrelationDict[name]=0
    return pearsonCorrelationDict
```

# Cosine Similarity
Normally the formula shown below is used to calculate the angule between two vectors, if the vectors are similar the angle is small and the value is close to 1, on the other hand if the vectors are not related at all the angle is close to 90 degrees (if we think the data as being 2d).

However, this formula as it is cannot be used since every user may have different standards on how the rate movies,this difference can cause an problem.Therefore, it is necessary to normalize the vectors( in this case only make the distribution have zero mean) by adjusting the formula.Modifying the equation will end up in a special case of Pearson correlation coefficient.

$ \frac{\sum \limits _{i=0} ^{n}(rating_i)(rating _j)} {\sqrt{\sum \limits _{i=0} ^{n}(rating_i)^2}\sqrt{\sum\limits _{i=0} ^{n}(rating_j)^2}} $


**Note Cosine similarity as the original formula show, will not work in this case( proved in experiments) since every users has a special way to rate movies, cosine similairy does not count for that (like pearson relation that normalizes the ratings for calculations)**

```python
def cosine_sim(user_group,input_df):
    cosine_sim={}
    for name, group in userSubsetGroup:

        group = group.sort_values(by='movieId')
        input_df = input_df.sort_values(by='movieId')
        #Get the N for the formula
        nRatings = len(group)
        #Get the review scores for the movies that they both have in common
        temp_df = input_df[input_df['movieId'].isin(group['movieId'].tolist())]
        #And then store them in a temporary buffer variable in a list format to facilitate future calculations
        tempRatingList = temp_df['rating'].tolist()
        #Let's also put the current user group reviews in a list format
        tempGroupList = group['rating'].tolist()
        user_average=temp_df['rating'].mean()
        crew_average=group['rating'].mean()
        num=sum((i)*(j) for i,j in zip(tempRatingList,tempGroupList))
        magnitude_x=sqrt(sum((i)**2 for i in tempRatingList))
        magnitude_y=sqrt(sum((i)**2 for i in tempGroupList))
        den=magnitude_x*magnitude_y

        if num!=0 and den!=0:
            cosine_sim[name] = num/den
        else:

            cosine_sim[name]=0
    return cosine_sim
```

Calculate similarity between users:

Here is where we get a list of simialr users where we calculate the pearson coefficient (similarity metric) and sort the users based on that.

At the end we get a dataframe consisting on similar users and the recommendation( where com_df is complete dataframe meaning some of the recommendations there may be already in the list of movies the user has watched which indicates the system recommend things the user already like and sub_df consisting on only recommendations the user has not seen)

```python
def collaborative_recommendation(pearsonCorrelationDict):
    pearsonDF = pd.DataFrame.from_dict(pearsonCorrelationDict, orient='index')
    pearsonDF.columns = ['similarityIndex']
    pearsonDF['userId'] = pearsonDF.index
    pearsonDF.index = range(len(pearsonDF))
    
    topUsers=pearsonDF.sort_values(by='similarityIndex', ascending=False)[0:50]
    topUsersRating=topUsers.merge(rating_df, left_on='userId', right_on='userId', how='inner')
   
    topUsersRating['weightedRating'] = topUsersRating['similarityIndex']*topUsersRating['rating']
    tempTopUsersRating = topUsersRating.groupby('movieId').sum()[['similarityIndex','weightedRating']]
    tempTopUsersRating.columns = ['sum_similarityIndex','sum_weightedRating']
    
    
    recommendation_df = pd.DataFrame()

    recommendation_df['weighted average recommendation score'] = tempTopUsersRating['sum_weightedRating']/tempTopUsersRating['sum_similarityIndex']
    recommendation_df['movieId'] = tempTopUsersRating.index
    recommendation_df = recommendation_df.sort_values(by='weighted average recommendation score', ascending=False)
    
    
    com_df=movies_df.loc[movies_df['movieId'].isin(recommendation_df.head(15)['movieId'].tolist())]
    
    sub_df=com_df.loc[~com_df['movieId'].isin(user['movieId'].tolist())]
    
    return sub_df,com_df,topUsersRating
```

### Content base

**User-Item**

In this approach rates and genres are combined to generate weigthed table in where preferences of the user are reflected.
Here how it works is as follow:
Input matrix of the form nxm:
where n is the number of movies watched(rows) and is the genres(columns):
$$\begin{bmatrix}
\hline 
Action & Drama & Comedy & Horror\\
 1 & 0 & 1 & 1\\
 1 & 0 & 0 & 1 \\
 1& 1 & 1 & 0 \\
\end{bmatrix}$$
We get the transpose of that matrix (Imagine we rotate the initial matrix 90 degres counterclokwise), now the rows are the genres and the columns are the movies, where a 1 represent that movie has that genre.
$$\begin{bmatrix}
\hline 
Action & 1 & 1 & 1\\
 Drama & 0 & 0 & 1\\
 Comedy & 1 & 0 & 1 \\
 Horror & 1 & 1 & 0 \\
\end{bmatrix}$$

At the end we get the dot product of the rating given to that movie with the whole row.

$$\begin{bmatrix}
\hline 
Action & 1*rating_x & 1*rating_y & 1*rating_z\\
 Drama & 0*rating_x & 0*rating_y & 1*rating_z\\
 Comedy & 1*rating_x & 0*rating_y & 1*rating_z \\
 Horror & 1*rating_x & 1*rating_y & 0*rating_z \\
\end{bmatrix}$$

We make a summation over the rows and then the resuting value is the genre weight.

That gives us an scalar per row( scalar means just a number) where the user preferences are reflected in that value.


```python
def content_base(df_movies,user,input_df):
    user_movies=df_movies[df_movies['movieId'].isin(user['movieId'])]
    user_movies.drop(['movieId','title'],axis=1,inplace=True)
    user_movies=user_movies.reset_index(drop=True)

    # user movies has the form of a matrix with nxm entries where n is the number of movies user has watched and m the total genres
    
    
    userProfile = user_movies.transpose().dot(input_df['rating'])
    
    userProfile=userProfile.sort_values(ascending=False)
    userProfile.plot(kind='bar',color=['C0', 'C1', 'C2','C3'],xlabel='genres',ylabel='Weights',title='User preferences')
    plt.show()
    genreTable = df_movies.set_index(df_movies['movieId'])
    genreTable.drop(['movieId','title'],axis=1,inplace=True)
    genreTable
    
    recommendationTable_df = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())
    recommendationTable_df=recommendationTable_df.sort_values(ascending=False)
    
    com=movies_df.loc[movies_df['movieId'].isin(recommendationTable_df.head(15).keys())]
    rec=com.loc[~com['movieId'].isin(user['movieId'].tolist())]
    
    return rec,com,userProfile
```

```python
content,content_sub,userProfile=content_base(df_movies,user,input_df)
```

Example:

By doing this the preferences of each user is taken into account at the time of training the model to predict ratings based on genres

Regressor Model

The task of the model is to predict what rating the user will give to any given recommendation.

In the training data, rating given by the user were used whether in the testing data( model has not seen this data) we will pretend that these data are the real recommendations, and since the user has not watch them it is necessary to transform this data differently.Instead of using user's ratings we will use the average rating that Similar users give to that movie.

Before we define what exactly is this model, let's explain briefly what is the goal. Here, we want rating the user will give to a movie given the input( the one we worked in the cells just above). Therefore, we are predicitng a number and in simple words that is a regression problem.

Now that we define the problem, how do we approach it. Based on several regressor models used( such as linear regression, DecisionTree Regressor etc) a RandomForestRegressor was selected.

The RandomForestRegressor is an ensemble model( meaning its ouptut consist on a combination of several small models) that combines several regular DecisionTree to obtain an output. One of the advantages of ensemble models is it will outperform the prediciotns of every contributing small model.

Now, for decision tree definition look at this link where is clearly explained: https://towardsdatascience.com/what-is-a-decision-tree-22975f00f3e1

Here the documentation of Decision tree and RandomForestRegressor:

* Decision tree: https://scikit-learn.org/stable/modules/tree.html
* RandomForest Regressor: https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html

**Something really important here:Every user have a random forest regressor that will predict their ratings, we could not train one model for all users since there was not a feature to indicate two users are different unless we include the user id as a feature,but then we would end up with 610 classes which at the end is not determinant to the modelso by giving one model per user we avoid that scenario.**

### Error
**The metric used to error is mean absolute error**

$ MAE= \frac{\sum \limits _{i=0} ^{n}|y_t-y_p|} {n}$

* y_t is the true value
* y_p is the predicted value
* n is the number of elements

**MAE is calculated for every user's predictions and at the end we can take the mean of all predictions to see the overall error for n users**

$ \frac{\sum \limits _{i=0} ^{n}MAE_i} {n} $


We can also see the error per user depending on the number of movies that user has watched using a scatter plot

```python
sns.scatterplot(x=collaborative_results['error'],y=collaborative_results['# movies'])
plt.title('Mean absolute error vs Number of movies watched')
plt.show()
```

On the other hand, we can also see that the number of movies watched and the error may not be realted at all although the few ones with few movies are the ones with highest error, that is not the general trend. Therefore, the MAE must be related to the number of movies watched AND other thing ( maybe the user does not have a consistent way to rate movies).

### Coverage
**Coverage is a metric used in recommendation systems to measure the amount of unique movies recommended to n given users**
**The purpose of this metric is to know if the catalog of movies recommended is big enough to cover all users likes, but also with meaningful recommendations**

$ Coverage= \frac{r} {n}$

where:
* r is unique items recommended
* n total items

### Personalization
Measures the correlation between all users recommendaions to determine how unique each set of recommendations is. If this value is low it means users are getting the same recommendations no matter user input.


We will calculate the correlation between every set of recommended movies, in other words how similar every user personal recommendations is.
Then, once we have the matrix, we will take the upper triangle and get the values there.
$$\begin{bmatrix} 1 & .3 & .3 \\ .5 & 1 & .1 \\ 0 & .5 & 1 \end{bmatrix}$$
We need to exclude the main diagnola because that is where the matrix is calculating item similarity with itself(and we are not interested on that), so we can generalize and just take the upper triangle.

# Conclusion

Base on the metrics provided above, we can conclude that collaborative filtering offers the best experience overall since it has a better personalization and great catalogue of recommendations.

On the other hand, content approach did not perfom as expected, its coverage is not that good and the personalization neither. The reason of this may be because the data is highly bias towards drama and comedy, affecting the algorithm to recommend those genres (since its is bias more users got those genres as favorites affecting the algorithm)

Synopsis approach perfom better than expected, it has good coverage and personalization( altough some randomess was added to the process of selecting movies to recommend that may improve the algorithm)

Popularity as expected has the lower coverage altough its personalization is better than the content approach.

The ratings prediciton were better than expected, initially I thought the prediction may have a margin error within 1 point, but the results were within a margin error of .35( .5 if we take increments of .5 between ratings).
